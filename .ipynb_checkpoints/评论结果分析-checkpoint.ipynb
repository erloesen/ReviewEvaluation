### 评价抽取比例

```python
def classify_count(c):
    if c == 0:
        return '0'
    elif c == 1:
        return '1'
    elif 2 <= c <= 4:
        return '2-4'
    elif 5 <= c <= 10:
        return '5-10'
    else:
        return '10+'
```

```python
merge_st['count_bin'] = merge_st['count'].apply(classify_count)
grouped = merge_st.groupby(['电视剧名称', 'count_bin']).size().reset_index(name='n')
total_per_drama = grouped.groupby('电视剧名称')['n'].transform('sum')
grouped['percent'] = grouped['n'] / total_per_drama * 100
```

```python
pivot_df = grouped.pivot(index='电视剧名称', columns='count_bin', values='percent').fillna(0)
pivot_df = pivot_df[['0', '1', '2-4', '5-10', '10+']]
colors = dict(zip(pivot_df.columns, palette))

fig, ax = plt.subplots(figsize=(12, 8))
bottom = pd.Series([0] * len(pivot_df), index=pivot_df.index)

for category in pivot_df.columns:
    heights = pivot_df[category]
    bars = ax.bar(pivot_df.index, heights, bottom=bottom, label=category, color=colors[category])

    # 添加百分比文本
    for i, (b, h) in enumerate(zip(bottom, heights)):
        if h > 1:  # 避免太小的百分比堆叠混乱
            ax.text(i, b + h / 2, f"{h:.1f}%", ha='center', va='center', fontsize=10)
    bottom += heights

# 格式设置
ax.set_ylabel("百分比 (%)")
ax.set_xlabel("电视剧名称")
ax.set_title("评价维度提取数量")
ax.legend(title="提取数量", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.savefig(f"./img/评价维度提取数量", dpi=300, bbox_inches='tight') 
```

```python
def cut_words_jieba(text):
    words = jieba.cut(text)
    use = [word for word in words if len(word) > 1 and word not in final_stopwords]
    return use
```

```python
def cut_words_jieba2(text):
    text = re.sub(r"[^\u4e00-\u9fa5]", "", text)
    # stop_pos = {'u', 'e', 'y', 'o', 'd', 't', 'f', 'p', 'c', 'r', 'm', 'q'}  # 词性缩写
    useful_flags = {'n', 'nr', 'ns', 'nt', 'nz', 'vn', 'v', 'vd', 'a', 'an', 'ad'}
    words = pseg.cut(text)
    use = [word for word, flag in words if flag in useful_flags and len(word) > 1 and word not in final_stopwords]
    return use
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(drama_feature.values)

# 聚类数范围（2~10）
K_range = range(2, 11)
sil_scores = []

# 迭代不同聚类数
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    sil_scores.append(score)

# 绘制轮廓系数
plt.figure(figsize=(8, 5))
plt.plot(K_range, sil_scores, marker='o')
plt.xlabel('聚类数 k')
plt.ylabel('轮廓系数 (Silhouette Score)')
plt.title('不同聚类数下的聚类效果')
plt.grid(True)
plt.tight_layout()
plt.show()

# 最佳聚类数
best_k = K_range[sil_scores.index(max(sil_scores))]
print(f"最佳聚类数 k = {best_k}, 轮廓系数 = {max(sil_scores):.3f}")
```

```python
# 重新聚类
kmeans_best = KMeans(n_clusters=best_k, random_state=42, n_init='auto')
drama_feature['cluster'] = kmeans_best.fit_predict(X_scaled)

# 输出每一类的剧集名称（行索引）
clustered = drama_feature.groupby('cluster')

for cluster_id, group in clustered:
    print(f"\n📺 类别 {cluster_id}：")
    print(group.index.tolist())
```

```python
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer
from collections import Counter
from umap import UMAP
from sklearn.feature_extraction.text import CountVectorizer
from bertopic.representation import KeyBERTInspired
from hdbscan import HDBSCAN

# 适合中文任务，性能和速度较优
embedding_model = SentenceTransformer("shibing624/text2vec-base-chinese")
```

```python
def jieba_tokenizer(text):
    return [w for w in jieba.lcut(text) if w not in final_stopwords and len(w) > 1]
```

```python
df_list = []

for (first, second, motion), group in wjdwj.groupby(['一级', '二级', '正负向']):
    # if second != '社会反思' or motion != '负向':
    #     continue
    count = len(group)
    print(second, motion, count)
    review = group['依据'].tolist()
    if count < 100:
        n_neighbors = min(10, max(3, count//5))
        min_cluster_size = min(5, max(2, count//20))
    else:
        n_neighbors = 20
        min_cluster_size = 15
    umap_model = UMAP(n_neighbors=n_neighbors, n_components=5, metric='cosine')
    hdbscan_model = HDBSCAN(min_cluster_size=min_cluster_size)
    topic_model = BERTopic(embedding_model=embedding_model, umap_model=umap_model, hdbscan_model=hdbscan_model, language="chinese") 
    topics, _ = topic_model.fit_transform(review)
    
    topic_info = topic_model.get_topic_info()
    topic_counts = Counter(topics)
    df = []
    for i, row in topic_info.iterrows():
        if row['Topic'] != -1:
            keywords = topic_model.get_topic(row['Topic'])
            count_topic = topic_counts.get(row['Topic'])
            for word, freq in keywords:
                df.append({'关键词':word.replace(" ", ""), '频数':freq*count_topic})
    df = pd.DataFrame.from_records(df)
    df = df.groupby('关键词')['频数'].sum().reset_index()
    df['二级'] = second
    df['情感'] = motion
    df_list.append(df)
final = pd.concat(df_list, axis=0)
final = final.sort_values('频数', ascending=False)
final
```

```python
import colorsys

def darken_color(rgb, factor=0.5):
    r, g, b = rgb
    h, l, s = colorsys.rgb_to_hls(r, g, b)
    l = max(0, l * factor)  # 减小亮度
    r2, g2, b2 = colorsys.hls_to_rgb(h, l, s)
    return f"#{int(r2*255):02x}{int(g2*255):02x}{int(b2*255):02x}"
```

```python
word_colors = {}

for i, ct in enumerate(['主题立意', '价值导向', '社会反思', '思想深度']):
    base_color = base_colors[i]
    r, g, b = (np.array(base_color)*255).astype(int)
    colorp = f"#{r:02x}{g:02x}{b:02x}"
    word_colors[(ct, '正向')] = colorp
    colorn = darken_color(base_color)
    word_colors[(ct, '负向')] = colorn
word_colors
```
