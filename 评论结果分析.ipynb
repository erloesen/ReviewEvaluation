### 评价抽取比例

```python
def classify_count(c):
    if c == 0:
        return '0'
    elif c == 1:
        return '1'
    elif 2 <= c <= 4:
        return '2-4'
    elif 5 <= c <= 10:
        return '5-10'
    else:
        return '10+'
```

```python
merge_st['count_bin'] = merge_st['count'].apply(classify_count)
grouped = merge_st.groupby(['电视剧名称', 'count_bin']).size().reset_index(name='n')
total_per_drama = grouped.groupby('电视剧名称')['n'].transform('sum')
grouped['percent'] = grouped['n'] / total_per_drama * 100
```

```python
pivot_df = grouped.pivot(index='电视剧名称', columns='count_bin', values='percent').fillna(0)
pivot_df = pivot_df[['0', '1', '2-4', '5-10', '10+']]
colors = dict(zip(pivot_df.columns, palette))

fig, ax = plt.subplots(figsize=(12, 8))
bottom = pd.Series([0] * len(pivot_df), index=pivot_df.index)

for category in pivot_df.columns:
    heights = pivot_df[category]
    bars = ax.bar(pivot_df.index, heights, bottom=bottom, label=category, color=colors[category])

    # 添加百分比文本
    for i, (b, h) in enumerate(zip(bottom, heights)):
        if h > 1:  # 避免太小的百分比堆叠混乱
            ax.text(i, b + h / 2, f"{h:.1f}%", ha='center', va='center', fontsize=10)
    bottom += heights

# 格式设置
ax.set_ylabel("百分比 (%)")
ax.set_xlabel("电视剧名称")
ax.set_title("评价维度提取数量")
ax.legend(title="提取数量", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.savefig(f"./img/评价维度提取数量", dpi=300, bbox_inches='tight') 
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(drama_feature.values)

# 聚类数范围（2~10）
K_range = range(2, 11)
sil_scores = []

# 迭代不同聚类数
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    sil_scores.append(score)

# 绘制轮廓系数
plt.figure(figsize=(8, 5))
plt.plot(K_range, sil_scores, marker='o')
plt.xlabel('聚类数 k')
plt.ylabel('轮廓系数 (Silhouette Score)')
plt.title('不同聚类数下的聚类效果')
plt.grid(True)
plt.tight_layout()
plt.show()

# 最佳聚类数
best_k = K_range[sil_scores.index(max(sil_scores))]
print(f"最佳聚类数 k = {best_k}, 轮廓系数 = {max(sil_scores):.3f}")
```

```python
# 重新聚类
kmeans_best = KMeans(n_clusters=best_k, random_state=42, n_init='auto')
drama_feature['cluster'] = kmeans_best.fit_predict(X_scaled)

# 输出每一类的剧集名称（行索引）
clustered = drama_feature.groupby('cluster')

for cluster_id, group in clustered:
    print(f"\n📺 类别 {cluster_id}：")
    print(group.index.tolist())
```
